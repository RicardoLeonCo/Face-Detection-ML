```python
#First we need to install the Open Source Computer Vision library
#in Python.

import cv2
```

```python
#This library comes with many pretrained classifiers for different
#parts of the face such as the eyes, smile, and more. We need to
#download the pre-trained face detection model from:

#Face-detection model: https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml

#We also need to save the file in our directory.
face_cascade = cv2.CascadeClassifier('face_detector')
```

```python
#Let's test our code by picking an image. In this case
#I picked this image of Taylor Swift and her famous celebrity
#friends, aka "Taylor's Squad."

img = cv2.imread('tsquad.png')
```

```python
#Face detection code using built-in functions.
#The method detects objects in the input image ("tsquad"), and returns
#the detected objects (faces) as a list of rectangles.

faces = face_cascade.detectMultiScale(img, 1.1, 4)
```

```python
#To wrap things up now we will draw rectangles around the detected
#faces and see the results.

for(x, y, w, h) in faces:
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2) #draws a rectangle outline using opposite corners
  #known as pt. 1 and pt. 2
cv2.imwrite("face_detected.png", img)
print("Successfully saved")
```

#**Notes on the Performance of the Model**

As we can see the performance of the model on this picture was somewhat good. There were many faces that it didn't detect and part of that was probably because in the picture the subjects are making funny faces which could make it harder for the model to detect their faces. Also it mistakenly confused a celebrity's legs for a face.
